The sine function is a trigonometric function which computes the ratio of the lenght of the side opposite to an angle to the lenght of the longest side of a rigth triangle. Denoted $ Sin(\theta) $ with $ \theta $ the angle.


We only need an angle to compute the sine function, hence a single input is required. Our quatum circuit will then be a single-qubit one and will run on a device of 1 wire.

Let's discretize our interval, $ $

```python
from sklearn.model_selection import train_test_split

# Sine function
def discretize_int(num_points):
    X = np.linspace(0, 2 * np.pi, num_points)
    np.random.shuffle(X)
    y = np.sin(X)
    
    return X, y


data_X, data_Y = discretize_int(1000)
# Split the data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, test_size=0.3, random_state=42)
```

The sine function is bounded in the interval [-1, 1] so we do not need to normalized our data or handle any pre-processing. there is not feature to learn from this data either, so no need for feature engineering.
